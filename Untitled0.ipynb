```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
df.head()
```

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)
```

```python
marker_cluster = MarkerCluster().add_to(restaurant_map)

for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']],
                  popup=row['Restaurant Name']).add_to(marker_cluster)
```

```python
restaurant_map.save('restaurant_distribution_map.html')
```

```python
grouped_df = df.groupby('City')
```

```python
concentration_map = folium.Map(location=map_center, zoom_start=12)

for city, city_df in grouped_df:
    city_marker = folium.Marker([city_df['Latitude'].mean(), city_df['Longitude'].mean()],
                                icon=folium.Icon(color='blue')).add_to(concentration_map)
```

```python
concentration_map.save('restaurant_concentration_map.html')
```

```python
average_ratings_by_city = grouped_df['Aggregate rating'].mean()
```

```python
cuisines_by_city = grouped_df['Cuisines'].apply(lambda x: ',').reset_index()
```

```python
price_ranges_by_city = grouped_df['Price range'].mean()
```

```python
print("Average Ratings by City:")
print(average_ratings_by_city)
```
